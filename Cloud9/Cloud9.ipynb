# Cloud 9

# What is Cloud9 ?

Simply put, Cloud 9 is a development environment on the cloud provided by amazon web services (AWS). 

Basically, it allows us to write, run and debbug code in a EC2 machine. It's been really helpful for me, because it allows me to work in a machine with powerful hardware, without expending a load of money on a personal computer. It's also possible to access it from anywhere, so I can log in at work, at home, from a friend's home and so on. 

So, in the current notebook, I'd like to show how to take the first steps towards using Cloud9.

# Set Up

After logging in AWS, we can go to the [Cloud9 homepage](https://aws.amazon.com/de/cloud9/) and select create environment.

![cloud9_homepage](Cloud9/cloud9_homepage.png)

Then, we are prompted to configure the environment.

![cloud9_config](Cloud9/cloud9_basic_configuration.png)

For our purposes here, I'll just go with the standard configurations. This will create a new free-tier eligible EC2 machine. 

However, if you already have an existing EC2 machine, you can select "Create and run in remote server (ssh connection)". This will prompt you to insert the public IP address of your EC2 machine and show you the public key for the ssh connection. This public key has to be added to the .ssh/authorized_keys of your already existing machine. Then you'll be good to go.

*Note, you might have to install nodejs in your machine, as Cloud9 requires it for connection*

Finally, AWS sets the environment for us:

![cloud9_setup](Cloud9/cloud9_setup.png)

# Checking the EC2 Instances

As I mentioned, AWS creates a EC2 machine, where our development environment is installed. We can check it by visiting the EC2 page.

# Connection to a Database

This now works just like any other machine. We can use it to connect to a database, also hosted in AWS, and write some records.

1) We have to download the psycopg2 package. This can be done via the bash shell at the bottom of the Cloud9 screen. This is necessary to connect to the postgres database we will be working with.

2) We can then run the script, which does the following:

    - It imports the required modules (including the credentials we need to connect to the database)
    
    - It generates some dummy data that will be inserted in a column that already exists in the database.
    
    - It writes the records to the table
    
    
*Note: for this to work, you have to make sure that the security group of your cloud hosted Database is accepting connections from cloud 9*
